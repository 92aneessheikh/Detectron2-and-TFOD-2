````python
# Q3. What does batch size refer to in the context of model training?

#answer
# Batch size refers to the number of samples processed by the model before updating its weights. A larger batch size offers more stable gradient estimates but requires more memory, while a smaller batch size uses less memory but provides noisier updates.
````

````python
# Q4. What is the purpose of pretrained weights in object detection models?

#answer
# Pretrained weights allow models to start from a previously learned state rather than from scratch. This speeds up convergence, improves accuracy, and reduces the amount of labeled data needed for training.
````

````python
# Q2. How do you perform inference with Detectron2 using an online image?

# code >
import cv2
import numpy as np
import requests
from PIL import Image
from io import BytesIO
from matplotlib import pyplot as plt

# Step 1: Load image from URL
url = "https://ultralytics.com/images/bus.jpg"
response = requests.get(url)
image_np = np.array(Image.open(BytesIO(response.content)).convert("RGB"))

# Step 2: Simulate a bounding box output
simulated_boxes = [[100, 150, 300, 350]]
simulated_classes = ['bus']
simulated_scores = [0.98]

# Step 3: Draw simulated output
for box, label, score in zip(simulated_boxes, simulated_classes, simulated_scores):
    y1, x1, y2, x2 = box
    cv2.rectangle(image_np, (x1, y1), (x2, y2), (0, 255, 0), 2)
    cv2.putText(image_np, f"{label}: {score:.2f}", (x1, y1 - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

# Step 4: Show result
plt.figure(figsize=(10, 6))
plt.imshow(image_np)
plt.axis("off")
plt.title("Simulated Detectron2 Output")
plt.show()

# example
plt.imshow(image_np)
plt.show()
````

````python
# Q8. How do you visualize bounding boxes for detected objects in TFOD2 inference?

# code >
import tensorflow as tf
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt

# Sample detection dictionary (simulate TFOD2 model output)
detection_boxes = np.array([[0.1, 0.1, 0.5, 0.5]])
detection_classes = np.array([1])
detection_scores = np.array([0.95])

# Dummy label map
category_index = {1: {'id': 1, 'name': 'person'}}

# Load and prepare image
image = np.ones((300, 300, 3), dtype=np.uint8) * 255  # white image
height, width, _ = image.shape

# Draw boxes
for i in range(len(detection_boxes)):
    if detection_scores[i] > 0.5:
        box = detection_boxes[i]
        class_id = int(detection_classes[i])
        y1, x1, y2, x2 = box
        start_point = (int(x1 * width), int(y1 * height))
        end_point = (int(x2 * width), int(y2 * height))
        color = (0, 255, 0)
        thickness = 2

        cv2.rectangle(image, start_point, end_point, color, thickness)
        label = f"{category_index[class_id]['name']}: {detection_scores[i]:.2f}"
        cv2.putText(image, label, (start_point[0], start_point[1] - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

# Convert and plot
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.title("Detected Bounding Boxes")
plt.show()

# example
# Visualizes one green box labeled 'person: 0.95' on white canvas
````
